```python
username = str(input("Dame el username:" )) 
password = str(input("Dame el password:"))
print("Username:", username)
print("Password:", password)
if len(username) >=5 and len(username) < 10:
    print(True)
else:
    print(False)

if password == 'Tokio' or password == 'Python':
    print(True)
else:
    print(False)
```
